## Encode data

### 1. Ordinal Encoding for Categorical Data


```python
# Define the custom ordinal encoding mapping
gender_mapping = {
    'Female': 0,
    'Male': 1
}

# learning_mode_mapping = {
#     'Face to Face': 1,
#     'Face to Face, Asynchronous Online Learning (On your own time)': 2,
#     'Face to Face, Synchronous Online Learning (Real Time)': 3,
#     'Face to Face, Synchronous Online Learning (Real Time), Asynchronous Online Learning (On your own time)': 4,
#     'Asynchronous Online Learning (On your own time)': 5,
#     'Synchronous Online Learning (Real Time)': 6,
#     'Synchronous Online Learning (Real Time), Asynchronous Online Learning (On your own time)': 7
# }

study_level_mapping = {
    'Certificate/Diploma': 1,
    'Undergraduate': 2,
    'Postgraduate': 3
}

income_mapping = {
#     'Less than RM 3000': 1,
    'Less than RM 4,849': 1,
#     'RM 3001 - 10 000': 3,
    'RM 4,850 - RM10,959': 2,
#     'RM 10 001 - 25 000': 5,
    'More than RM10,960': 3
}

preference_mapping = {
    'Not at All': 0,
    'Not Really': 0,
    'Undecided': 0,
    'Somewhat': 0,
    'Very Much': 1
}

vak_mapping = {
    'Visual': 1,
    'Auditory': 2,
    'Kinesthetic': 3
}

# final_tools = {
#     '6. Online Instructional Strategies/Assessment [Demonstration]': 1,
#     '6. Online Instructional Strategies/Assessment [Digital Lab Experiments]': 2,
#     '6. Online Instructional Strategies/Assessment [Forum]':3,
#     '6. Online Instructional Strategies/Assessment [Case Study]': 4,
#     '6. Online Instructional Strategies/Assessment [Concept Mapping]': 5,
#     '6. Online Instructional Strategies/Assessment [Real Time Online Exam]':6,
#     '6. Online Instructional Strategies/Assessment [Individual Project/Assignment]': 7,
#     '6. Online Instructional Strategies/Assessment [Group Project/Assignment]': 8,
#     '6. Online Instructional Strategies/Assessment [Online Quiz/Test - MCQ]': 9,
#     '6. Online Instructional Strategies/Assessment [Online Quiz/Test - Essay]': 10,
#     '6. Online Instructional Strategies/Assessment [Online Quiz/Test - Open Book]': 11,
#     '6. Online Instructional Strategies/Assessment [Peer Review Assessment Live Presentation]': 12,
#     '6. Online Instructional Strategies/Assessment [Recorded Presentation]': 13

# }
```

```python
df['Gender'] = df['Gender'].map(gender_mapping)
df['Level of Study'] = df['Level of Study'].map(study_level_mapping)
df['Household Income'] = df['Household Income'].map(income_mapping)
df['Dominant_VAK'] = df['Dominant_VAK'].map(vak_mapping)
# df['final_assessment_tool'] = df['final_assessment_tool'].map(final_tools)
```

```python
# Check for null values in the DataFrame
null_values = df.isnull().sum()

# Display the null values
print("Null Values in the DataFrame:")
print(null_values)
```

```python
# Extract online assessment columns
online_asssessment_columns = df.iloc[:, 7:20]
online_asssessment_columns.head()
```

```python
for column in online_asssessment_columns.columns:
    df[column] = df[column].map(preference_mapping)
    
df.head()
```

### 2. One-hot encoding

```python
# One-hot encoding for multiple columns
columns_to_encode = ['Preferred learning mode', 'Preferred Social Media Platform', 'Preferred Communication Platform',
                    'Difficulties in Online Learning']
df = pd.get_dummies(df, columns=columns_to_encode, prefix=columns_to_encode)

df.head()
```
