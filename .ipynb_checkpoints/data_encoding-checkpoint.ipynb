```python
import pandas as pd
import numpy as np
```

## Load data

```python
df = pd.read_csv('Datasets/responses_dominantVAK.csv')

df
```

```python
distinct_values = df['Household Income'].unique()

for value in distinct_values:
    print(value)
```

```python
# Create a new column that contains the names of columns with 'Very Much' in the corresponding row
df['final_assessment_tool'] = df.apply(lambda row: '; '.join(col for col, cell in row.items() if cell == 'Very Much'), axis=1)

# Fill 'NA' for rows where there is no 'Very Much'
# df['final_assessment_tool'].replace('', 'NA', inplace=True)

# Replace empty strings with NaN
df['final_assessment_tool'].replace('', np.nan, inplace=True)

# Replace NaN values with column names where 'Somewhat' is selected
somewhat_columns = df.apply(lambda row: '; '.join(col for col, cell in row.items() if cell == 'Somewhat'), axis=1)
df['final_assessment_tool'].fillna(somewhat_columns, inplace=True)

df.head()
```

```python
df['final_assessment_tool'] = df['final_assessment_tool'].str.split('; ')

df = df.explode('final_assessment_tool').reset_index(drop = True)

df.head()
```

```python
distinct_values = df['final_assessment_tool'].unique()

for value in distinct_values:
    print(value)
```

## Encode data

### 1. Ordinal Encoding for Categorical Data


```python
# Define the custom ordinal encoding mapping
gender_mapping = {
    'Female': 0,
    'Male': 1
}

# learning_mode_mapping = {
#     'Face to Face': 1,
#     'Face to Face, Asynchronous Online Learning (On your own time)': 2,
#     'Face to Face, Synchronous Online Learning (Real Time)': 3,
#     'Face to Face, Synchronous Online Learning (Real Time), Asynchronous Online Learning (On your own time)': 4,
#     'Asynchronous Online Learning (On your own time)': 5,
#     'Synchronous Online Learning (Real Time)': 6,
#     'Synchronous Online Learning (Real Time), Asynchronous Online Learning (On your own time)': 7
# }

study_level_mapping = {
    'Certificate/Diploma': 1,
    'Undergraduate': 2,
    'Postgraduate': 3
}

income_mapping = {
#     'Less than RM 3000': 1,
    'Less than RM 4,849': 1,
#     'RM 3001 - 10 000': 3,
    'RM 4,850 - RM10,959': 2,
#     'RM 10 001 - 25 000': 5,
    'More than RM10,960': 3
}

preference_mapping = {
    'Not at All': 0,
    'Not Really': 0,
    'Undecided': 0,
    'Somewhat': 0,
    'Very Much': 1
}

vak_mapping = {
    'Visual': 1,
    'Auditory': 2,
    'Kinesthetic': 3
}

final_tools = {
    '6. Online Instructional Strategies/Assessment [Demonstration]': 1,
6. Online Instructional Strategies/Assessment [Written assignment]
6. Online Instructional Strategies/Assessment [Digital Lab Experiments]
6. Online Instructional Strategies/Assessment [Live Group Discussions]
6. Online Instructional Strategies/Assessment [Forum]
6. Online Instructional Strategies/Assessment [Case Study]
6. Online Instructional Strategies/Assessment [Journaling]
6. Online Instructional Strategies/Assessment [Video based feedback]
6. Online Instructional Strategies/Assessment [Live Chat]
6. Online Instructional Strategies/Assessment [Brainstorming]
6. Online Instructional Strategies/Assessment [Online Debate]
6. Online Instructional Strategies/Assessment [Blog Writing]
6. Online Instructional Strategies/Assessment [Problem Based Learning]
6. Online Instructional Strategies/Assessment [Internet Research]
6. Online Instructional Strategies/Assessment [Exit tickets (End Class Feedback)]
6. Online Instructional Strategies/Assessment [Learning Reflection]
6. Online Instructional Strategies/Assessment [Online Exercise/Practice]
6. Online Instructional Strategies/Assessment [Procedural Demonstration (Perform a given action)]
6. Online Instructional Strategies/Assessment [Concept Mapping]
6. Online Instructional Strategies/Assessment [Document Analysis]
6. Online Instructional Strategies/Assessment [Debugging]
6. Online Instructional Strategies/Assessment [Review (Article, Play, Movie, Audio, Book, etc.)]
6. Online Instructional Strategies/Assessment [Real Time Online Exam]
6. Online Instructional Strategies/Assessment [Individual Project/Assignment]
6. Online Instructional Strategies/Assessment [Group Project/Assignment]
6. Online Instructional Strategies/Assessment [Online Quiz/Test - MCQ]
6. Online Instructional Strategies/Assessment [Online Quiz/Test - Essay]
6. Online Instructional Strategies/Assessment [Online Quiz/Test - Open Book]
6. Online Instructional Strategies/Assessment [Peer Review Assessment Live Presentation]
6. Online Instructional Strategies/Assessment [Recorded Presentation]
6. Online Instructional Strategies/Assessment [Portfolio]
}
```

```python
df['Gender'] = df['Gender'].map(gender_mapping)
df['Level of Study'] = df['Level of Study'].map(study_level_mapping)
df['Household Income'] = df['Household Income'].map(income_mapping)
df['Dominant_VAK'] = df['Dominant_VAK'].map(vak_mapping)
```

```python
# Extract online assessment columns
online_asssessment_columns = df.iloc[:, 7:38]
online_asssessment_columns.head()
```

```python
for column in online_asssessment_columns.columns:
    df[column] = df[column].map(preference_mapping)
    
df.head()
```

### 2. One-hot encoding

```python
# One-hot encoding for multiple columns
columns_to_encode = ['Preferred learning mode', 'Preferred Social Media Platform', 'Preferred Communication Platform',
                    'Difficulties in Online Learning']
df = pd.get_dummies(df, columns=columns_to_encode, prefix=columns_to_encode)

df.head()
```
