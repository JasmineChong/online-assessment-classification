```python
import pandas as pd
```

## Load data

```python
df = pd.read_csv('Datasets/responses_dominantVAK.csv')

df.head()
```

```python
distinct_values = df['Household Income'].unique()

for value in distinct_values:
    print(value)
```

## Encode data

### 1. Ordinal Encoding for Categorical Data


```python
# Define the custom ordinal encoding mapping
gender_mapping = {
    'Female': 0,
    'Male': 1
}

# learning_mode_mapping = {
#     'Face to Face': 1,
#     'Face to Face, Asynchronous Online Learning (On your own time)': 2,
#     'Face to Face, Synchronous Online Learning (Real Time)': 3,
#     'Face to Face, Synchronous Online Learning (Real Time), Asynchronous Online Learning (On your own time)': 4,
#     'Asynchronous Online Learning (On your own time)': 5,
#     'Synchronous Online Learning (Real Time)': 6,
#     'Synchronous Online Learning (Real Time), Asynchronous Online Learning (On your own time)': 7
# }

study_level_mapping = {
    'Certificate/Diploma': 1,
    'Undergraduate': 2,
    'Postgraduate': 3
}

income_mapping = {
#     'Less than RM 3000': 1,
    'Less than RM 4,849': 1,
#     'RM 3001 - 10 000': 3,
    'RM 4,850 - RM10,959': 2,
#     'RM 10 001 - 25 000': 5,
    'More than RM10,960': 3
}

preference_mapping = {
    'Not at All': 0,
    'Not Really': 0,
    'Undecided': 0,
    'Somewhat': 0,
    'Very Much': 1
}

vak_mapping = {
    'Visual': 1,
    'Auditory': 2,
    'Kinesthetic': 3
}
```

```python
df['Gender'] = df['Gender'].map(gender_mapping)
df['Level of Study'] = df['Level of Study'].map(study_level_mapping)
df['Household Income'] = df['Household Income'].map(income_mapping)
df['Dominant_VAK'] = df['Dominant_VAK'].map(vak_mapping)
```

```python
# Extract online assessment columns
online_asssessment_columns = df.iloc[:, 7:38]
online_asssessment_columns.head()
```

```python
for column in online_asssessment_columns.columns:
    df[column] = df[column].map(preference_mapping)
    
df.head()
```

### 2. One-hot encoding

```python
# One-hot encoding for multiple columns
columns_to_encode = ['Preferred learning mode', 'Preferred Social Media Platform', 'Preferred Communication Platform',
                    'Difficulties in Online Learning']
df = pd.get_dummies(df, columns=columns_to_encode, prefix=columns_to_encode)

df.head()
```

## Save into a new csv file

```python
# Save into a new csv
df.to_csv('Datasets/responses_encoded.csv', index=False)
```
